```python
#Importing the basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
vehicle_capacity=70
No_of_Vehicles=10
vehicles = [i for i in range(1, 5)]
```

```python
#Create a set for each of the customers and the depot and between the customers for distance calculation
A = [(i, j) for i in dataset.CUSTOMER for j in dataset.CUSTOMER if i != j]
```

```python
#Measuring the distances between each points
c = {(i, j): np.hypot(dataset.XCOORD[i]-dataset.XCOORD[j], dataset.YCOORD[i]-dataset.YCOORD[j]) for i, j in A}
```

```python
c
```

```python
sorted(c)
```

```python
#Importing the CPLEX Model
from docplex.mp.model import Model
```

```python
model = Model('CVRP')
```

```python
#Parsing parameters
x = model.binary_var_dict(A, name='x')
u = model.continuous_var_dict(dataset.CUSTOMER[1:], ub=vehicle_capacity, name='u')
v= model.continuous_var_dict(vehicles, ub=No_of_Vehicles, name='y')
```

```python
#Parsing objectives and constraints
model.minimize(model.sum(c[i, j]*x[i, j] for i, j in A))
model.add_constraints(model.sum(x[i, j] for j in dataset.CUSTOMER if j != i) == 1 for i in dataset.CUSTOMER[1:])
model.add_constraints(model.sum(x[i, j] for i in dataset.CUSTOMER if i != j) == 1 for j in dataset.CUSTOMER[1:])
model.add_indicator_constraints(model.indicator_constraint(x[i, j], u[i]+dataset.DEMAND[i] == dataset.DEMAND[j]) for i, j in A if i != 0 and j != 0)
model.add_constraints(u[i] >= dataset.DEMAND[i] for i in dataset.CUSTOMER[1:])
model.add_constraints(u[i] >= dataset.DEMAND[i] for i in dataset.CUSTOMER[1:])
model.parameters.timelimit = 60
solution = model.solve(log_output=True)
```

```python
#Display only optimal solution lines
active_arcs = [a for a in A if x[a].solution_value > 0.9]
```

```python
plt.figure(figsize=(15,10))
plt.scatter(dataset.XCOORD[1:], dataset.YCOORD[1:], c='b')
for i in dataset.CUSTOMER[1:]:
    plt.annotate('$q_%d=%d$' % (i, dataset.DEMAND[i]), (dataset.XCOORD[i]+0.5, dataset.YCOORD[i]))
for i, j in active_arcs:
    plt.plot([dataset.XCOORD[i], dataset.XCOORD[j]], [dataset.YCOORD[i], dataset.YCOORD[j]], c='g', alpha=0.3)
plt.plot(dataset.XCOORD[0], dataset.YCOORD[0], c='r', marker='^')
```
